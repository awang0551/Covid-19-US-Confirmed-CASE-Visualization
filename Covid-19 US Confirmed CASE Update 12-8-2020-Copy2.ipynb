```python
import numpy as py
import pandas as pd
import plotly as plt
from numpy.random import randn
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
```

```python
df=pd.read_csv('covid_19_data.csv')
```

```python
df=pd.DataFrame(df) 
```

```python
df
```

```python
df=df.rename(columns={'Country/Region' : 'Country'})
```

```python
df=df.rename(columns={'ObservationDate':'Date'})
```

```python
df=df.rename(columns={'Province/State':'State'})
```

```python
df_us=df[df['Country'] == 'US']
```

```python
df_us.head(10)
```

```python
df_us_states = df_us.groupby(['State','Date']).sum().reset_index().sort_values('Confirmed',ascending=False)
```

```python
df_us_states.head(55)
```

```python
df_state_code = pd.read_csv('2011_US_AGRI_Exports')
```

```python
df_state_code
```

```python
df_state_code.columns
```

```python
df_state_code.drop(columns=['category', 'total exports', 'beef', 'pork', 'poultry',
       'dairy', 'fruits fresh', 'fruits proc', 'total fruits', 'veggies fresh',
       'veggies proc', 'total veggies', 'corn', 'wheat', 'cotton', 'text'],inplace=True)
```

```python
print(df_state_code)
```

```python
df_state_code=df_state_code.rename(columns={'state':'State'})
```

```python
df_us_code=pd.merge(df_us_states, df_state_code, on='State')
```

```python
print(df_us_code)
```

```python
#california is missing..??
len(df_us_code['code'].unique())
```

```python
df_us_code = df_us_code.groupby(['code','Date']).sum().reset_index().sort_values('Confirmed',ascending=False)
```

```python
df_us_code = df_us_code[df_us_code['Confirmed']>0]
df_us_code = df_us_code.groupby(['Date','code']).sum().reset_index()
print(df_us_code)
```

```python
px.choropleth(df_us_code, 
              locations = df_us_code['code'],
              color=df_us_code['Confirmed'], 
              animation_frame=df_us_code['Date'],
              color_continuous_scale = 'Portland',
              locationmode='USA-states',
              scope="usa",
              range_color=(0, 1800000))
```
